```matlab
t = linspace(0, 6*pi, 200);
x = cos(t);
y = sin(t);
z = 4*t;
T = numel(t);
N = 10;

s = 0.1;
x = x + s * randn(N, T);
y = y + s * randn(N, T);
z = z + s * randn(N, T);

figSizeScale([3 3])
plot3(x', y', z');
view([45 20])
```

```matlab
size(x)
```

```matlab
trials = {cat(3, x, y, z)};
trials
size(t)
```

```matlab
world = Teetool.build_world(t, trials, resolution=[10 10 10]);
```

```matlab
means = Teetool.get_means(world);
```

```matlab
plot3(means.xvec, means.yvec, means.zvec)
```
